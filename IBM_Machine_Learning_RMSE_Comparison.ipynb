```python
import matplotlib.pyplot as plt
import numpy as np


models = ['KNN', 'NMF', 'NN Embedding LR', 'NN Embedding Ridge', 'NN Embedding Lasso', 'NN Embedding Elastic Net']
rmse_values = [1.0182, 1.3697, 0.8174, 0.8157, 0.8155, 0.8155]  # RMSE values for each model

# Create a bar plot
plt.figure(figsize=(10, 6))
colors = ['blue', 'orange', 'green', 'red', 'purple', 'cyan']  # Different colors for each bar
plt.bar(models, rmse_values, color=colors)

plt.xlabel('Models', fontsize=12)
plt.ylabel('RMSE', fontsize=12)
plt.title('RMSE Comparison of Different Models', fontsize=14)
plt.ylim(0, max(rmse_values) + 0.2)  # Add some space above the highest bar

# Annotate each bar with its RMSE value
for i, v in enumerate(rmse_values):
    plt.text(i, v + 0.02, f'{v:.4f}', ha='center', va='bottom', fontsize=10)

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=15)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping
plt.show()
```
